### 4.3 Công thức tính elo

Công thức để cập nhật elo theo từng trận: (tham khảo từ https://www.ergosum.co/nate-silvers-nba-elo-algorithm/ )<br>

$R_i$: Elo of a team in $i^{th}$ match

$R_{i + 1} = k * (S_{team} - E_{team} + R_i)$

$S_{team} = 1$: win || $0$: lose

$E_{team} = \frac{1}{1 + 10^{\frac{oppElo - teamElo}{400}}}$

$k = 20 \frac{(MOV_{winner} + 3)^{0.8}}{7.5 + 0.006(\text{Elo_diff}_{winner})}$

$MOV_{winner} = winner_{score} - loser_{score}$

$\text{Elo_diff}_{winner} = winner_{elo} - loser_{elo}$

```python
def K(MOV, elo_diff):
    f = 1 if MOV > 0 else -1 #factor
    k = 20 * (f * MOV + 3) ** (0.8) / (7.5 + 0.006 * (f * elo_diff))
    return k, k # win: +elo, lose: -elo

def elo_prediction(home_elo,away_elo):
    E_home = 1./(1 + 10 ** ((away_elo - home_elo) / (400.)))
    return E_home

def elo_update(H_dif, home_elo, away_elo):
    home_advan = 100. #home advantage
    home_elo += home_advan
    E_home = elo_prediction(home_elo, away_elo)
    E_away = 1 - E_home
    elo_diff = home_elo - away_elo
    MOV = H_dif # margin of victory
    
    S_home, S_away = (1, 0) if MOV > 0 else (0, 1)
    K_home, K_away = K(MOV, elo_diff)
        
    return K_home * (S_home - E_home), K_away * (S_away - E_away)

def get_pre_elo(df, index, team_name):
    while index < len(df) - 1:
        index += 1
        if (team_name == df.iloc[index]['H_Team']):
            H_pre_update, A_pre_update  = elo_update(df.iloc[index]['H_Dif'], df.iloc[index]['H_preElo'], df.iloc[index]['A_preElo'])
            return df.iloc[index]['H_preElo'] + H_pre_update
        elif (team_name == df.iloc[index]['A_Team']):
            H_pre_update, A_pre_update  = elo_update(df.iloc[index]['H_Dif'], df.iloc[index]['H_preElo'], df.iloc[index]['A_preElo'])
            return df.iloc[index]['A_preElo'] + A_pre_update

    # khi không tìm được trong dataframe, dùng elo của "preseason"
    return int(dict_pre_elo[team_name])
```

### 4.4 Tính các chỉ số trung bình của 5 trận gần nhất

```python
def get_recent(df, index, team_name, num_recent_games = 5):
    data = []
    
    while(len(data) < num_recent_games and index < len(df)-1):
        index += 1
        if (team_name == df.iloc[index]['H_Team']):
            data.append(list(df.iloc[index][lst_H]))
        elif (team_name == df.iloc[index]['A_Team']):
            data.append(list(df.iloc[index][lst_A]))
            
    if len(data) < num_recent_games:
        data.append(list(pre_stat_df.loc[pre_stat_df['Team'] == team_name].iloc[0]))
            
    return pd.DataFrame(data, columns = col_name).agg(['mean']).values.tolist()[0]
```

### 4.5 Xử lý dữ liệu

#### 4.5.1 Tính toán 2 cột H_elo và A_elo

```python
raw_df['H_preElo'], raw_df['A_preElo'] = np.zeros(len(raw_df)), np.zeros(len(raw_df))

for index in range(len(raw_df.index) - 1, -1, -1):
    H_name = raw_df.iloc[index]['H_Team']
    A_name = raw_df.iloc[index]['A_Team']
    
    raw_df.at[index, 'H_preElo'] =  round(get_pre_elo(raw_df, index, H_name), 1)
    raw_df.at[index, 'A_preElo'] =  round(get_pre_elo(raw_df, index, A_name), 1)
```

#### 4.5.2 Tính toán các giá trị trung bình cho từng trận và lưu vào dataframe mới

```python
data = []

for index in range(len(raw_df.index)):
    row = [raw_df.iloc[index]['Date']]
    
    H_name = raw_df.iloc[index]['H_Team']
    row += [H_name]
    row += get_recent(raw_df, index, H_name)
    row += [raw_df.iloc[index]['H_preElo']]
    
    A_name = raw_df.iloc[index]['A_Team']
    row += [A_name]
    row += get_recent(raw_df, index, A_name)
    row += [raw_df.iloc[index]['A_preElo']]
    
    row += [raw_df.iloc[index]['H_Dif']]
    data.append(row)
```

```python
lst_H_ave = ['H_Team'] + ['H_ave' + col for col in col_name[1:]] + ['H_preElo']
lst_A_ave = ['A_Team'] + ['A_ave' + col for col in col_name[1:]] + ['A_preElo']
lst_ave_cols = lst_new_cols = ['Date'] + lst_H_ave + lst_A_ave + ['H_Dif']
data_df = pd.DataFrame(data, columns=lst_new_cols)
data_df.head()
```

### 4.6 Lưu dữ liệu đã được xử lý

File <b>'data.csv'</b>: Chứa thông tin (thời gian, tên 2 đội, chỉ số trung bình của cả 2, elo trước trận và kết quả) của từng trận <br>
File <b>'description.txt.'</b>: Chứa ý nghĩa các tên cột mới (ví dụ: H_ave2P: số quả ném 2 điểm trung bình trong 5 trận của đội Home)

```python
def export_final_data():
    data_df.to_csv('../Data/data.csv', encoding='utf-8', index=False)
    
    with open('../Data/raw_description.txt', 'r') as f:
        desc = {'preElo':'Elo before the match\n'}
        for i, line in enumerate(f.readlines()):
            c, d = line.split(' : ')
            if i:
                desc[c[2:]] = d[5:]
            else:
                desc[c] = d

    with open('../Data/description.txt', 'w') as f:
        f.write('Date'+ ' : ' + desc['Date'])
        for col in lst_new_cols[1:-1]:
            h_str = 'Home ' if 'H_' in col else 'Away '
            ave_str = 'Average ' if 'ave' in col else ''

            main = col.replace('H_', '').replace('A_', '').replace('ave', '')

            f.write(col + ' : ' + h_str + ave_str + desc[main])

        f.write('H_Dif'+ ' : ' + 'Difference between home score and away score (H_Dif > 0: Home win)')

export_final_data()
```

# 7. Huấn luyện model và dự đoán kết quả

3. Huấn luyện model với <b>Neural Network (MLPCLassifier)</b> _ sử dụng <b>RandomizedSearchCV</b> vì chạy nhanh hơn nhưng vẫn đảm bảo kết quả

### 7.1 Khám phá dữ liệu tập huấn luyện

```python
train_X_df.dtypes
```

Có 3 cột không phải dạng numeric: <b>Date, H_Team, A_Team</b> => không cần xem sự phân bố của chúng

```python
df = train_X_df.select_dtypes(exclude=['object'])
df.agg([missing_percentage, 'min', median, 'max'])
```

### 7.2 Tiền xử lý tập huấn luyện

Xóa các cột 'object'

```python
def remove_object_col(X):
    return X.select_dtypes(exclude=['object'])
```

```python
preprocess_pipeline = make_pipeline(FunctionTransformer(remove_object_col), StandardScaler())
preprocess_pipeline
```

Kết quả trả về sẽ là <b>train_X_df</b> sau khi đã tiền xử lý, là một mảng Numpy

```python
preprocess_pipeline.fit_transform(train_X_df)
```

### 7.3 Huấn luyện model với Logistic Regression

```python
logistic_regression_model = LogisticRegression(random_state=0)

LRparam_grid = {
    'logisticregression__C': [0.001, 0.01, 0.1, 1, 10, 100, 1000], # Inverse of regularization strength
    'logisticregression__max_iter': list(range(100,1000,100)), # Maximum number of iterations taken for the solvers to converge.
    'logisticregression__solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'] # Algorithm to use in the optimization problem
}

full_pipeline = make_pipeline(FunctionTransformer(remove_object_col),
                              StandardScaler(), 
                              logistic_regression_model)

logistic_regression_gs = GridSearchCV(estimator = full_pipeline, 
                                      param_grid = LRparam_grid, 
                                      scoring = 'accuracy', 
                                      cv = 5, # cross-validation
                                      return_train_score=True)
logistic_regression_gs
```

```python
logistic_regression_gs.fit(train_X_df, train_y_sr)
logistic_regression_gs.best_params_
```

---

### 7.4 Huấn luyện model với Random Forest

```python
random_forest_model = RandomForestClassifier(random_state=0,
                                             oob_score = True,
                                             min_samples_leaf = 50)

RFparam_grid = {
    'randomforestclassifier__n_estimators': list(range(100,1000,200)), # Number of trees in random forest
    'randomforestclassifier__max_features': ['auto', 'sqrt', 'log2'], # Number of features to consider at every split
    'randomforestclassifier__min_samples_leaf': list(range(10,100,10)), # Min number of samples required to split an internal node
}

full_pipeline = make_pipeline(FunctionTransformer(remove_object_col),
                              StandardScaler(),
                              random_forest_model)

random_forest_rs = RandomizedSearchCV(estimator = full_pipeline, 
                                      param_distributions = RFparam_grid, 
                                      scoring = 'accuracy', 
                                      cv = 5,
                                      return_train_score = True,
                                      n_iter = 100, # random n_iter
                                      n_jobs = -1) # use all processors
random_forest_rs
```

```python
random_forest_rs.fit(train_X_df, train_y_sr)
random_forest_rs.best_params_
```

---

### 7.5 Huấn luyện model với Neural Network

```python
neural_network_model = MLPClassifier(random_state=0)

NNparam_grid = {
    'mlpclassifier__max_iter': [100, 300, 500, 1000],
    'mlpclassifier__hidden_layer_sizes':np.arange(10, 15),
    'mlpclassifier__alpha': [0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50, 100],
}

full_pipeline = make_pipeline(FunctionTransformer(remove_object_col),
                              StandardScaler(), 
                              neural_network_model)

neural_network_rs = GridSearchCV(estimator = full_pipeline, 
                                  param_grid = NNparam_grid, 
                                  scoring = 'accuracy', 
                                  cv = 5,
                                  return_train_score = True,
                                  n_jobs = -1) # use all processors

neural_network_rs
```

```python
neural_network_rs.fit(train_X_df, train_y_sr)
neural_network_rs.best_params_
```

### 7.7 Thử nghiệm thực tế với lượt trận đầu tiên của mùa giải sắp tới

<b>Bước 1:</b> Lấy ngày thi đấu và tên 2 đội trong lượt trận đầu tiên của mùa giải (2021-2022) <br>
Ở lượt trận có 2 bảng đấu là Eastern và Western, mỗi bảng có 15 đội => Trong lượt đấu 1 mỗi bảng có 7 trận => tổng 14 trận

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2022_games.html'

page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
tbl = soup.find("table")
temp_df = pd.read_html(str(tbl))[0]
temp_df = temp_df[['Date', 'Visitor/Neutral', 'Home/Neutral']]
temp_df.head()
```

<b>Bước 2:</b> Gộp với dataframe cũ raw_df

```python
# Đổi tên cột của temp_df
temp_df = temp_df.rename(columns={'Visitor/Neutral':'A_Team', 'Home/Neutral':'H_Team'})
# Đổi thành tên viết tắt trong temp_df
temp_df['H_Team'] = temp_df.apply(lambda x: getShortName(x['H_Team']), axis=1)
temp_df['A_Team'] = temp_df.apply(lambda x: getShortName(x['A_Team']), axis=1)
temp_df.head()
```

```python
new_raw_df = pd.DataFrame(temp_df[:14].append(raw_df), columns=raw_df.columns)
new_raw_df = new_raw_df.reset_index(drop=True)
new_raw_df.shape
```

Cập nhật elo trước trận cho lượt trận đầu tiên (14 dòng đầu)

```python
for index in range(14 - 1, -1, -1):
    H_name = new_raw_df.iloc[index]['H_Team']
    A_name = new_raw_df.iloc[index]['A_Team']
    
    new_raw_df.at[index, 'H_preElo'] = round(get_pre_elo(new_raw_df, index, H_name), 1)
    new_raw_df.at[index, 'A_preElo'] = round(get_pre_elo(new_raw_df, index, A_name), 1)

new_raw_df.head()
```

Tính toán các chỉ số trung bình ở 5 lượt trận gần nhất của 2 đội trong 14 dòng đầu

```python
data = []

for index in range(14):
    row = [new_raw_df.iloc[index]['Date']]
    
    H_name = new_raw_df.iloc[index]['H_Team']
    row += [H_name]
    row += get_recent(new_raw_df, index, H_name)
    row += [new_raw_df.iloc[index]['H_preElo']]
    
    A_name = new_raw_df.iloc[index]['A_Team']
    row += [A_name]
    row += get_recent(new_raw_df, index, A_name)
    row += [new_raw_df.iloc[index]['A_preElo']]
    
    row += [new_raw_df.iloc[index]['H_Dif']]
    data.append(row)
```

```python
new_data_df = pd.DataFrame(data, columns = data_df.columns)
new_data_df.head()
```

```python
new_data_df.shape
```

Nhiệm vụ bây giờ mình cần làm là dự đoán kết quả ở cột Output của trong <b>new_data_df</b> (cột 'H_Win')

```python
# Lấy các cột input
new_test_X_df = new_data_df.drop("H_Win", axis=1)
```

Sử dụng mô hình có điểm số cao nhất là <b>Neural Network (MLPClassifier)</b> để dự đoán kết quả vòng 1 giải bóng rổ NBA

```python
results = neural_network_rs.predict(new_test_X_df)
for i in range(14):
    date = new_data_df.iloc[i]['Date']
    H_team = new_data_df.iloc[i]['H_Team']
    A_team = new_data_df.iloc[i]['A_Team']
    result = 'win' if results[i] else 'lose'
    
    print("%s: %s will %s %s" % (date, H_team, result, A_team))
```
