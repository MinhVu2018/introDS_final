## 1. Parse HTML

```python
year = 2000
main_url = 'https://www.basketball-reference.com'
year_url = f'/leagues/NBA_{year}_games.html'

sleep_time = 15 # Đơn vị: giây
```

```python
requests_cache.install_cache()
```

```python
def get_html(url):
    # r = requests.get(url)
    # if r.from_cache == False:
    #     print('Not from cache')
    #     time.sleep(sleep_time)
    # else: print('From cache')

    with webdriver.Chrome('chromedriver',chrome_options=chrome_options) as driver:
        driver.get(url)
        
        # Creating object of an Actions class
        action = ActionChains(driver)
        ele = driver.find_element_by_class_name('section_heading_text')
        # Performing the mouse hover action on the target element.
        action.move_to_element(ele).perform()
        button = driver.find_element_by_xpath('//*[@class="section_heading_text"]//*[text()="Get table as CSV (for Excel)"]')
        button.click()

        html_text = driver.page_source

    tree = BeautifulSoup(html_text, 'html.parser')
    return tree

tree = get_html(main_url + year_url)
```

### Get all data from one season

```python
filter = tree.find('div', class_=['filter'])
for i, tag in enumerate(filter.find_all('a')):
    if i: # first link same with main page // no need to get html
        link = main_url + tag['href']
        print(link)
        tree = get_html(link)

    # print(tree.find('pre', class_=['csv_schedule']))

    break
```
